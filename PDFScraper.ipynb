```python
import tabula
import warnings
import plotly.express as px
from datetime import datetime
from google.colab import drive, files, output # specific to Google Colab

warnings.filterwarnings('ignore')
```

```python
url="https://assets.publishing.service.gov.uk/government/uploads/system/uploads/attachment_data/file/287937/07-1184x-tackling-obesities-future-choices-report.pdf"
```

```python
table=tabula.read_pdf(url, pages=45)
```

```python
table[0]
```

```python
df = table[0]
df.dropna(subset=df.columns[3:6], inplace=True)
df
```

```python
import pandas as pd
```

```python
df1=df
```

```python
df1.at[9, 'Unnamed: 0'] = 'NHS costs attributable to obesity alone'
df1.at[12, 'Unnamed: 0'] = 'Wider total costs of overweight and obesity (7x direct costs)'
```

```python
df2=df1
df2.rename(columns={'Unnamed: 0': 'Measure'}, inplace=True)
```

```python
df2
```

```python
df3=df2.drop(['2007','2015'],axis=1)
```

```python
df3
```

I only want my data to be in billions, so will drop the % row

```python
df4 = df3.drop([13])
```

```python
df4
```

```python
df4.to_csv('Future costs of elevated BMI (£ billion per year).csv', index=False)
```
